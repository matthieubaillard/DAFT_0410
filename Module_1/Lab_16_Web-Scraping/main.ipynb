````python
#Method 1

#tags = ['h1', 'p'] # this includes h1 and/or p that are not the developers list itself (such as titles or repository names)
#developers = [element.text.strip() for element in soup.find_all(tags)]
#print('\n'.join(developers))

#print('\n\n\n')

#Method 2
#dev_tags = soup.find_all('h1', {'class': 'h3 lh-condensed'}) # this returns the actual developers list, but not their respsctive user name
#dev_names = [tag.find('a').text.strip() for tag in dev_tags]
#print('\n'.join(dev_names))

# Each developer is within its own <article class="Box-row d-flex"> in the original html code.
# Find all <article> tags with class "Box-row d-flex"

developers = soup.find_all('article', class_='Box-row d-flex')
#print(developers)

#Inside each article, we want to retrieve the text of the links inside h1 tags with a specific class: <h1 class="h3 lh-condensed"><a>Developper name we want  to retrieve</a></h1>
#and the text inside the <p class="f4 text-normal mb-1"><a class="Link--secondary">User name we want to retrieve</a></p>
#and list them in the following format:

#['Developer name 1 (user name 1)',
# 'Developer name 2 (user name 2)',
# ...]

developers_list = []

# iterate over the developers and extract the desired information

for developer in developers:

    # extract the developer name from the <a> tag inside the <h1> tag with a h3 lh-condensed class:
    developer_name = developer.find('h1', class_='h3 lh-condensed').find('a').text.strip()

    #extracting user name if it exists (we need to catch the error before running the text.strip method in case there is no user name)
    #here we cannot run the find function twice as we did with the developer names, as the second one would return an error in case where the developers do not display their user name (AttributeError: 'NoneType' object has no attribute 'find')
    #select_one method to search for the 1st occurrence of an <a class="Link--secondary"> that is a descendant of a <p class="f4 text-normal mb-1">
    user_name_find = developer.select_one('p.f4.text-normal.mb-1 a.Link--secondary')
    user_name = user_name_find.text.strip() if user_name_find else None
    
    # append the result to the results list
    if user_name is not None:
        developers_list.append(f'{developer_name} ({user_name})')
    else:
        developers_list.append(developer_name)

developers_list
````

#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
response = requests.get(url).content
sp = BeautifulSoup(response,'html.parser')
````

````python
repo_selector = 'h2 > a'
len(sp.select(repo_selector))
````

#### Display all the image links from Walt Disney wikipedia page.

````python
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
response = requests.get(url).content
sp = BeautifulSoup(response,'html.parser')
````

````python
# Find all links that have a class = 'image'

img_href = sp.find_all("a", class_="image") 

#Iterate through the set of links to retrieve their href attribute, and recreate the absolute path with replace()

for a in img_href:
    print(a['href'].replace("/wiki", "https://en.wikipedia.org/wiki"))
````

#### Find the number of titles that have changed in the United States Code since its last release point.

````python
#Titles in bold have been changed since the last release point.
#Those are in <div> tags with class = 'usctitlechanged'

changed_titles = sp.find_all("div", class_="usctitlechanged") 

print(len(changed_titles))

#37 titles have changed in the US Code since its last release point.
````
