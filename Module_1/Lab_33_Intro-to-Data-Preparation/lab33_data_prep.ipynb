```python
#Turn the values of 'TransactionID', ''ClientID', 'BirthYear' and 'Department' to integers

df = df.astype({'TransactionID': int, 'ClientID': int, 'Department': int})

#'BirthYear': int -- let's deal with it later as there are missing values
df
```

```python
#Check if any duplicated on TransactionID and ClientID
transaction_duplicates = df[df.duplicated(subset='TransactionID', keep=False)]
print(transaction_duplicates)
#We see that there are no TransactionID duplicates

client_duplicates = df[df.duplicated(subset='ClientID', keep=False)]
print(client_duplicates.sort_values(by='ClientID'))
#Clients can have several transactions, so there are duplicates
#Checking them here actually helps us find some missing or inconsistent data

```

```python
#Fill up missing BirthYear = 1967 for ClientID = 34987 and BirthYear = 1999 for ClientID = 34988
df.loc[(df['ClientID'] == 34987) & (df['BirthYear'].isna()), 'BirthYear'] = 1967
df.loc[(df['ClientID'] == 34988) & (df['BirthYear'].isna()), 'BirthYear'] = 1999

#We were wondering if sailer stood for sailor or salesperson, but we can see that that Client is actually a (sales?) Manager.
#Fill up missing Profession = BDM for Client = 35008
df.loc[(df['ClientID'] == 35008) & (df['Profession'].isna()), 'Profession'] = 'BDM'
df.loc[(df['ClientID'] == 34987), 'Profession'] = 'Manager'

df
```
