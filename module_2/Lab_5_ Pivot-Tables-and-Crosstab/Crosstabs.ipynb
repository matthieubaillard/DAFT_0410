### 2. Use crosstabs to show the number of moovies per country

```python
moovies_country = pd.crosstab(index=moovies['country'], columns='number of movies')

#sort by number of movies, then by country's alphabetical order

moovies_country = moovies_country.sort_values(['number of movies', 'country'], ascending=[False, True])

moovies_country



# NB: TESTS TO DROP THE COL_0 INDEX FAILED
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
#Number of movies per director:

moovies_director = pd.crosstab(index=moovies['director_name'], columns='number of movies')

#sort by number of movies, then by country's alphabetical order

moovies_director = moovies_director.sort_values(['number of movies', 'director_name'], ascending=[False, True])

moovies_director
```

```python
#Number of movies per genres and country per director (sorted by number of movies)

pd.crosstab(index=[moovies['director_name'], moovies['country'], moovies['genres']], columns='number of movies').sort_values(['number of movies', 'director_name'], ascending=[False, True])
```

```python
#Number of movies per country per genre

moovies_genre = pd.crosstab(index=moovies['genres'], columns = moovies['country'])
moovies_genre
```

```python
#Number of movies per country per genre

moovies_genre = pd.crosstab(index=[moovies['genres'], moovies['country']], columns = 'number of films')
moovies_genre
```

```python
#Number of movies per director per genre per country (sorted by country)

pd.crosstab(index=[moovies['country'], moovies['genres'], moovies['director_name']], columns='number of movies').sort_values(['country', 'genres', 'director_name'], ascending=[True, True, True])
```

### 4. Average duration of films grouped by color

```python
pd.crosstab(index=moovies['color'], columns='average duration', values=moovies['duration'], aggfunc=[np.mean])
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
director_film_likes = pd.crosstab(index=moovies['director_name'], columns='average of cumulated FB likes of main actors per film', values=moovies[['actor_1_facebook_likes', 'actor_2_facebook_likes', 'actor_3_facebook_likes']].sum(axis=1), aggfunc=np.mean)
director_film_likes = director_film_likes.sort_values(['average of cumulated FB likes of main actors per film'], ascending=False)
director_film_likes
```
